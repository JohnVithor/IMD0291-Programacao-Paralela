```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rand = pd.read_csv("PI - rand().csv")
rand_r = pd.read_csv("PI - rand_r(%seed).csv")
```

```python
rand_r.index = rand_r.loc[:, "Tempos"]
rand_r = rand_r.rename_axis("Tamanho do Problema")
rand_r = rand_r.replace({',': '.'}, regex=True)
rand_r = rand_r.astype(float)
speedup = rand_r.iloc[:,8:14]
eficiencia = rand_r.iloc[:,14:]
```

```python
speedup.columns = [2, 3, 4, 6, 8, 12]
speedup = speedup.transpose()
speedup
```

```python
max_cores = 12
ax = speedup.plot()
plt.xlim(0, max_cores+1)
plt.ylim(0, max_cores+1)
plt.plot(range(max_cores+2), range(max_cores+2), label="Identidade", color='red', linestyle='--')
plt.legend(loc='best')
plt.xlabel(xlabel='Cores')
plt.ylabel(ylabel='Speedup')
plt.show()

```

```python
eficiencia.columns = [2, 3, 4, 6, 8, 12]
eficiencia = eficiencia.transpose()
eficiencia

```

```python
ax = eficiencia.plot()
plt.legend(loc='best')
plt.xlabel(xlabel='Cores')
plt.ylabel(ylabel='Eficincia')
```
